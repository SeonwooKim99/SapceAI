```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = 'D:/AI_Alignment/data_for_AI/data1163/'
before = 'wfm.txt'
rnn = 'wfm_after_alignment(RNN).txt'
cnn = 'wfm_after_alignment.txt'
```

```python
wfm_list=[]
with open(path+before, 'r',encoding = 'utf-16') as file:
    lines = file.readlines()
    wfm_data = np.array([list(map(float, line.split())) for line in lines[16:]])
    wfm_list.append(wfm_data)

plt.imshow(wfm_data,extent=[0, 300, 0, 300], origin='lower')  
plt.colorbar().ax.set_title('Wave', pad=10)
plt.xlabel('X axis [mm]')
plt.ylabel('Y axis [mm]')
plt.title('Before Algorithm')
plt.show()
```

```python
wfm_list=[]
with open(path+rnn, 'r',encoding = 'utf-16') as file:
    lines = file.readlines()
    wfm_data = np.array([list(map(float, line.split())) for line in lines[16:]])
    wfm_list.append(wfm_data)

plt.imshow(wfm_data,extent=[0, 300, 0, 300], origin='lower')  
plt.colorbar().ax.set_title('Wave', pad=10)
plt.xlabel('X axis [mm]')
plt.ylabel('Y axis [mm]')
plt.title('After RNN')
plt.show()
```
